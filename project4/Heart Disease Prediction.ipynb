# scaling data

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
df[columns_to_scale] = standardScaler.fit_transform(df[columns_to_scale])
```

# data training

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=99,test_size=0.33)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(criterion='gini',
                           max_depth=8,
                           min_samples_split=10,
                           random_state=5
                          )
```

```python
clf.fit(X_train,y_train)
```

```python
#important features
clf.feature_importances_
```

```python
df.columns
```

# make predictions

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

# confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#Check for overfitting and underfitting
print('Training set score:',clf.score(X_train, y_train))
print('Test set score:',clf.score(X_test, y_test))
```

# cross validation

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(clf,X_train,y_train,cv=10)
scores
```

```python
print('Average cross-validation score: {:.4f}'.format(scores.mean()))
```

Using the mean cross-validation, we can conclude that we expect the model to be around 95.05 % accurate on average.

# classification report

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

# ROC-AUC

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
# Plot the ROC curve
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2,label='ROC')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--',label='Random Guess')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
# compute ROC AUC

from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_pred)

print('ROC AUC : {:.4f}'.format(ROC_AUC))
```

The ROC AUC (Receiver Operating Characteristic Area Under the Curve) serves as a concise metric summarizing the performance of a classifier. As the value approaches 1, the classifier's performance improves.

Given that the ROC AUC of our model is nearing 1, we can confidently infer that our classifier excels in predicting heart disease.

```python
#plot of importance features
features=df.columns
importances=clf.feature_importances_
indices=np.argsort(importances)
```

```python
plt.title('feature importances')
plt.barh(range(len(indices)),importances[indices],color='b',align='center')
plt.yticks(range(len(indices)),[features[i] for i in indices])
plt.xlabel('features')
```
