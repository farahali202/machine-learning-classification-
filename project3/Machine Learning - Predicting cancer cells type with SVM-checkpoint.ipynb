# Data Visualization 

```python
#correlation between var
correlation = df.corr()
```

```python
correlation['Class'].sort_values(ascending=False)
```

The correlation coefficient, which ranges from -1 to +1, serves as a measure of the strength and direction of linear relationships between variables.

A correlation coefficient close to +1 indicates a strong positive correlation. In this dataset, we observe a robust positive correlation between 'Class' and variables such as 'Bare_Nuclei,' 'Uniformity_Cell_Shape,' and 'Uniformity_Cell_Size.'

Conversely, a correlation coefficient close to -1 suggests a strong negative correlation. When the coefficient approaches 0, it indicates little to no correlation. Notably, all variables exhibit positive correlations with the 'Class' variable. Some variables demonstrate a strong positive correlation, while others display a negative correlation.

```python
#correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation,annot=True,cmap='Blues')
```

'Class' exhibits a high positive correlation (correlation coefficient = 0.82) with 'Uniformity_Cell_Size,' 'Uniformity_Cell_Shape,' and 'Bare_Nuclei.'

'Class' displays positive correlations with 'Clump_thickness' (correlation coefficient=0.72), 'Marginal_Adhesion' (correlation coefficient=0.70), 'Single_Epithelial_Cell_Size' (correlation coefficient = 0.68), and 'Normal_Nucleoli' (correlation coefficient=0.71).

'Class' shows a weaker positive correlation with 'Mitoses' (correlation coefficient=0.42).

'Mitoses' exhibits weak positive correlations (correlation coefficient < 0.50) with all other variables in the dataset.


# Training model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_tra,y,random_state=4,test_size=0.2)
X_train.shape
X_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
param_grid = [
    {'C': [0.1, 1, 10, 100, 1000],
     'gamma': [0.0001, 0.001, 0.01, 0.1, 1],
     'kernel': ['rbf']},
    {'C': [0.1, 1, 10, 100, 1000],
     'kernel': ['linear']}
]
grid = GridSearchCV(SVC(), param_grid, verbose=2)
grid.fit(X_train, y_train)
classifier=grid
```

```python
grid.best_params_
```

```python
# predicting 
y_pred=grid.predict(X_test)
```

# Evaluation(results)

```python
from sklearn import metrics
print('confusion matrix:\n', metrics.confusion_matrix(y_test, y_pred))
print('accuracy:', np.round(metrics.accuracy_score(y_test, y_pred), 4))
print('precision:', np.round(metrics.precision_score(y_test, y_pred), 4))
print('recall:', np.round(metrics.recall_score(y_test, y_pred), 4))
print('f1-score:', np.round(metrics.f1_score(y_test, y_pred), 4))
print(metrics.classification_report(y_test, y_pred))
```
